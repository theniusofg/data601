Scrape the information from the following website: https://www.scrapethissite.com/pages/simple/ 
show the values in a dataframe

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://www.scrapethissite.com/pages/simple/')
response.ok
```

```python
response.text[:1000]
```

```python
html_doc = response.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
t = soup.find_all('div', {"class":"col-md-4 country"})
t
```

```python
country = [country.h3.text.strip() for country in t]
capital = [capital.find('span', {'class':'country-capital'}).text for capital in t]
pop = [population.find('span', {'class':'country-population'}).text for population in t]
area = [area.find('span', {'class':'country-area'}).text for area in t]
```

```python
import pandas as pd
df = pd.DataFrame()
df['country'] = country
df['capital'] = capital
df['population'] = pop
df['area'] = area
df
```
