```python
# Use this or replace this with yours from your week10 hw. (It's up to you.)

import pandas as pd
import random
from faker import Faker

class FakeDataGenerator:
    
    def __init__(self, df = None, rows=5):
        """
        This class generates fake data
        
        If data frame is passed in it will add the fake data to it. 
        Otherwise, it will create a new data frame
        
        Parameters:
        -----------
        df: 
          DataFrame default is None 
        rows:
          If df None, this parameter will be used to create a data frame with `row` length
        """
        # code goes under this line. 
        # I will deduct points if any code is written above the comments!
                                
        if type(df) == type(None):
            self.df = pd.DataFrame(index=range(0,rows)) 
        else:
            self.df = df
                
        self.rows = self.df.shape[0]
        
        
    def add_random_number(self, column_name, small_value, high_value): # optional: add probabilities 
        """
        adds fake numbers to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().random_int(min= small_value, max=high_value) for row in range(self.rows)]
        return self.df       
    
        
    def add_first_name(self, column_name):
        """
        adds fake first name to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().first_name() for row in range(self.rows)]
        return self.df
    
    
    def add_last_name(self, column_name):
        """
        adds fake last name to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().last_name() for row in range(self.rows)]

        return self.df
    
    
    def add_email(self, column_name):
        """
        adds fake email to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().email() for row in range(self.rows)]
        return self.df
    
    
    def add_phone_number(self, column_name):
        """
        adds fake phone number to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().phone_number() for row in range(self.rows)]
        return self.df
    
    
    def add_address(self, column_name):
        """
        adds fake address to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().address() for row in range(self.rows)]
        return self.df
```

```python
# complete this
class PocDataGenerator(FakeDataGenerator):

    def year_born(self, column_name):
        """
        adds fake year born to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [str(Faker().date_of_birth(minimum_age=17, maximum_age=70))[:4] for row in range(self.rows)]
        return self.df

    def num_children(self, column_name):
        """
        adds fake address to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().random_int(0,5) for row in range(self.rows)]
        return self.df

    def is_married(self, column_name):
        """
        adds fake marital status to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        self.df[column_name] = [Faker().boolean(48.2) for row in range(self.rows)]
        return self.df

    def annual_income(self, column_name):
        """
        adds annual income to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        

        self.df[column_name] = [Faker().random_elements(
            [(random.randint(12000,40000), 0.05),
                (random.randint(40000,70000), 0.45),
                (random.randint(70000,120000), 0.25),
                (random.randint(120000,250000), 0.125),
                (random.randint(250000,750000), 0.0625),
                (random.randint(750000,2000000), 0.0625)], 1, True)[0][0] 
                for row in range(self.rows)]
        return self.df

    def csv(self, file_name):
        """
        writes dataframe to csv
        
        Returns
        -------
        csv
        """
        self.df.to_csv(file_name)
```
